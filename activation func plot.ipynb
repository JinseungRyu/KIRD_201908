```python
import matplotlib.pylab as plt
import numpy as np

def sigmoid(x):
  return 1 / (1+np.exp(-x))

def stepFunc(x):
  return np.array(x>0, dtype = np.int)

def ReLU(x):
  return np.maximum(0,x)

def leakyReLU(x):
  return np.maximum(0.1*x, x)

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.figure(figsize=(12,6))

plt.subplot(121)
plt.plot(x,y, label = 'sigmoid')
plt.ylim(-0.1, 1.1)
y = stepFunc(x)
plt.plot(x,y, label = 'step', linestyle = '--')
y = ReLU(x)
plt.plot(x,y, label = 'ReLU', linestyle = '-.')
plt.legend()

plt.subplot(122)
y = np.tanh(x)
plt.plot(x,y, label = 'tanh')
plt.ylim(-1.1, 1.1)

y = leakyReLU(x)
plt.plot(x,y, label = 'leakyReLU', linestyle = '--')

plt.legend()
plt.suptitle("Activation Functions")

plt.show()
```
