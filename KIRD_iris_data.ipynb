## Iris 데이터: 3가지 종류의 붓꼿 데이터가 50개씩 순서대로 저장되어 있음


```
from sklearn import datasets
iris = datasets.load_iris()
 
print(iris.feature_names)
print(iris.target_names)
```

```
print(iris.data[:5])
print(iris.target[:5])
```

```
print(iris.data[100:105])
print(iris.target[100:105])
```

### np - 칼럼 합치기

```
merged = np.c_[iris.data, iris.target]
print(merged.shape)
print(merged[:10])
```

### numpy에서 칼럼간 correlation 구하기


```
np.corrcoef(merged, rowvar=False) # rowvar=False로 해야 칼럼간 관계구함. 아니면 행간의 관계구함.
```

## pandas DataFrame으로 변환하기

```
df = pd.DataFrame(np.c_[iris.data, iris.target])
df.describe()
```

### pandas에서 칼럼간 correlation 구하기




```
df.corr()
```
