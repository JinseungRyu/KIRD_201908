### 소프트웨어 개발자 vs 데이터과학자

### 소프트웨어 개발자적인 프로그램

```python
import random

position = 0
walk = [position]

for i in range(1000):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)
```

```python
walk[:20]
```

### 데이터 과학자

```python
승패 = np.random.randint(0, 2, size=1000)
```

```python
steps = np.where(승패>0, 1, -1)
```

```python
walk = steps.cumsum()
```

```python
walk
```

```python
%%timeit
for _ in range(5000):
    position = 0
    walk = [position]

    for i in range(1000):
        step = 1 if random.randint(0,1) else -1
        position += step
        walk.append(position)
```

```python
%%timeit
승패 = np.random.randint(0, 2, size=(5000,1000))
steps = np.where(승패>0, 1, -1)
walk = steps.cumsum(axis=1)
```
