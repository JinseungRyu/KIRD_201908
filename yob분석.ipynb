## 이름의 다양성을 어떻게 볼 수 있을까?

```python
비율합계 = 연도별_상위이름.pivot_table('비율', aggfunc='sum', index='연도', columns=['성별']).fillna(0)
```

```python
group=list(연도별_성별)[0][1]
```

```python
group=group.sort_values('비율',ascending=False)
```

```python
비율=group['비율']
```

```python
(비율.cumsum() < 0.5).sum()
```

```python
def 상위50이름갯수(group) :
    비율 = group['비율'].sort_values(ascending=False)
    return (비율.cumsum()<0.5).sum()
```

```python
상위50이름갯수(group)
```

```python
연도별_성별.apply(상위50이름갯수)[:10]
```

```python
연도별_상위50이름갯수 = 연도별_성별.apply(상위50이름갯수).unstack()
```

```python
연도별_상위50이름갯수.plot()
```
